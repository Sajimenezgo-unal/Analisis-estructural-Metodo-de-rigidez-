$$
\begin{equation}
    G_{xx}(x'_E,{\xi}'_E) =
    \begin{cases}
    G_{xx}^I(x'_E,{\xi}'_E) = \dfrac{L_E}{A_E E_E}{{\psi}_4}^E(x'_E){{\psi}_1}^E(\xi'_E) & 0 \leq x'_E < \xi'_E \\ 
    \\
    G_{xx}^{II}(x'_E,{\xi}'_E) = \dfrac{L_E}{A_E E_E}{{\psi}_1}^E(x'_E){{\psi}_4}^E(\xi'_E) & {\xi}'_E  {\leq} x'_E <  L_E
    \end{cases}
\end{equation}
$$
$$
\\
$$
$$
\begin{equation}
    G_{yy}(x'_E,{\xi}'_E) =
    \begin{cases}
    G_{yy}^I(x'_E,{\xi}'_E) = \dfrac{L^3_E}{6E_E I_E} \left[-\left(\dfrac{x'_E}{L_E}\right)^3{{\psi}_2}^E(\xi'_E) + 3\left(\dfrac{x'_E}{L_E}\right)^2 \frac{{{\psi}_3}^E(\xi'_E)}{L_E}\right]  & 0 \leq x'_E \leq {\xi}'_E \\ \\
    G_{yy}^{II}(x'_E,{\xi}'_E) = \dfrac{L^3_E}{6E_E I_E} \left[-\left(1-\dfrac{x'_E}{L_E}\right)^3{{\psi}_5}^E(\xi'_E) + 3\left(1-\dfrac{x'_E}{L_E}\right)^2 \dfrac{{{\psi}_3}^E(\xi'_E)}{L_E}\right]  & {\xi}'_E < x'_E \leq L_E 
    \end{cases}
\end{equation}
$$

### Transformación cargas externas 

    Cargas2Locales(dirR, Q, theta)

Se transforman las cargas externas de coordenadas globales a coordenadas locales, donde este método o formula recibe como variables los datos de inclinación del elemento en sentido antihorario (theta), la carga (Q) a transformar y la direción en la que sta horientada la carga en dirección de los ejes locales ya sea horizontal o vetical por lo que en este campo solo se recibe ya sea v, V para vertical o h, H para horizontal de lo contrartio arrojara un error.

y como salida entregara una lista con las cargas p y q en este mismo orden transformadas según correspondan.

Para el caso de la fuerza externa en dirección vertical $R'(x'_E)$, las cargas externas en coordenadas locales son:
$$
\begin{align}
    p_E(x'_E) &= R'(x'_E)|\cos{\theta_E}|\sin{\theta_E} \\
    q_E(x'_E) &= R'(x'_E)|\cos{\theta_E}|\cos{\theta_E} 
\end{align}
$$
<br />

De forma similar al caso de la fuerza externa en dirección vertical $R'(x'_E)$, las cargas externas en coordenadas locales son para una carga en dirección horizontal son:

<br />

$$
\begin{align}
        p_E(x'_E) =& Q'(x'_E)|\sin{\theta_E}|\cos{\theta_E}\\
        q_E(x'_E) =& -Q'(x'_E)|\sin{\theta_E}|\sin{\theta_E}
\end{align}
$$
